```python
# import the moons dataset
from sklearn.datasets import make_moons
```

```python
# split the dataset into features and targets
x_moons, y_moons = make_moons(n_samples=1000, noise=0.4)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_moons, y_moons, test_size=0.2, random_state=42)
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# create the pca pipeline
pca_pipeline = make_pipeline(StandardScaler(), PCA())
```

```python
# fit the pca pipeline
x_train_pca = pca_pipeline.fit_transform(x_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
# create the model params
params = {"max_depth":list(range(1,7)), "min_samples_split":[2,3,4], "max_leaf_nodes": list(range(2,100)),"min_samples_leaf":[1,2,3,4,5]}
```

```python
# Initialize the grid search
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid=params, scoring="accuracy", cv=10)
```

```python
# fit the grid search
grid_search.fit(x_train_pca, y_train)
```

```python
# check the best estimator
grid_search.best_estimator_
```

```python
# check the best score
grid_search.best_score_
```

```python
from sklearn.metrics import accuracy_score
```

```python
# check the test score
y_pred = grid_search.predict(pca_pipeline.transform(x_test))
accuracy_score(y_test, y_pred)
```
